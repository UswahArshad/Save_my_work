### Step 3. Assign it to a variable called users.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"

df = pd.read_csv(url, sep='|', index_col="user_id")

df.head()
```

### Step 4. Discover what is the mean age per occupation

```python
df.groupby(by="occupation")[["age"]].mean().add_prefix("MeanOf_").reset_index()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
out1 = df.groupby(by=["occupation", "gender"])[
    ["gender"]].count().add_prefix("CountOf_").reset_index()


out2 = out1.pivot(index="occupation", columns="gender",
                  values="CountOf_gender")


out2 = out2.fillna(0)

out1
out2
```

```python
res1 = out2.F + out2.M
res2 = out2.M/res1
final_res = res2.sort_values(ascending=False)
final_res
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
temp1 = df.groupby(by="occupation")[["age"]].min().add_prefix("Min_")
temp2 = df.groupby(by="occupation")[["age"]].max().add_prefix("Max_")
out5a = pd.concat([temp1, temp2], axis=1)
out5a
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
out5 = df.groupby(by=["occupation", "gender"])[
    ["age"]].mean().add_prefix("MeanOf_").reset_index()


out6 = out5.pivot(index="occupation", columns="gender", values="MeanOf_age")
out6
```
